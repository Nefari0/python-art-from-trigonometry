```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
log spiral combines the two functions:
cos(t)
e**(-t)

_____

x = cos(t)e**(t/k)
y = sin(t)e**(t/k)

k is set as a negative number

when k is large, fraction goes to 0
```

```python
# experimental
t = np.linspace(-2*np.pi,2*np.pi,1000)

k = -.01

x = np.cos(t)*np.exp(t/k)
y = np.sin(t)*np.exp(t/k)

plt.plot(x,y)
```

```python
# instructed
t = np.linspace(0,10*np.pi,1234)

k = -3

x = np.cos(t) * np.exp(t/k)
y = np.sin(t) * np.exp(t/k)

plt.plot(x,y,'r')
plt.axis('square')
plt.axis('off')

plt.show()
```

```python
# exercise.independent
t = np.linspace(0,10*np.pi,1000)

k = np.linspace(-5,-1,100)
clr = np.linspace(.01,1,len(k)) # color

for i in range(len(k)):

    x = np.cos(t) * np.exp(t/k[i])
    y = np.sin(t) * np.exp(t/k[i])

    plt.plot(x,y,color=[clr[i],.02,clr[i],],linewidth=2)

plt.axis('square')
plt.axis('off')

plt.show()
```

```python
# exercise.instructed
N = 150
ks = np.linspace(-6,-2,N)

for i in range(N):
    x = np.cos(t) * np.exp(t/ks[i])
    y = np.sin(t) * np.exp(t/ks[i])

    plt.plot(x,y,color=[i/N,0,i/N])
    
plt.axis('square')
plt.axis('off')

plt.show()
```
