```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rose curve example:

x = cos(kt)cos(t)
y = cos(kt)sin(t)

# t is time vector
# k is single number

if k is even integer, curve will have 2k pedals
if k is odd integer, curve will have k pedals
```

```python
# experimental
k = 2
t = np.linspace(0,2*np.pi,100)
# t = np.linspace(-1,1,100)

x = np.cos(k*t)*np.cos(t)
y = np.cos(k*t)*np.sin(t)

plt.plot(x,y)

```

```python
# instructed
k = 6.4

t = np.linspace(0,8*np.pi,1000)

x = np.cos(k*t) * np.cos(t)
y = np.cos(k*t) * np.sin(t)

plt.plot(x,y,'m',linewidth=3)
plt.axis('square')
plt.axis('off')
plt.show()
```

```python
# exercise.independent
fig,ax = plt.subplots(3,3,figsize=(10,6))
row, col = np.indices((3,3))

k = np.linspace(0,4,9)

t = np.linspace(0,8*np.pi,1000)

for i in range(9):
    
    x = np.cos(k[i]*t) * np.cos(t)
    y = np.cos(k[i]*t) * np.sin(t)
    
    r = row.ravel()[i]
    c = col.ravel()[i]
    
    ax[r,c].plot(x,y)
    ax[r,c].axis('off')
    ax[r,c].axis('square')
    
#     plt.plot(x,y)
    
plt.show()
```

```python
# exercise.instructed
fig,ax = plt.subplots(3,3,figsize=(10,6))
row,col = np.indices((3,3))

for i in range(0,9):
    
    k = i/2
    
    x = np.cos(k*t) * np.cos(t)
    y = np.cos(k*t) * np.sin(t)
    
    r = row.ravel()[i]
    c = col.ravel()[i]
    
    ax[r,c].plot(x,y,'b')
    ax[r,c].axis('off')
    ax[r,c].axis('square')
    ax[r,c].set_title('k=%s'%k)
#     print(k)
```
