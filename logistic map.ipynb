```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display,Math
```

##### ( _ indicates subscript)
##### x_(n+1) = rx_(n)(1 - x_n)

##### x_1 = .5
##### n E natural numbers : n <= 500
##### r E [1,4] in 1000 steps

```python

#----allows viewing, navigating plot---
# %matplotlib notebook 
#----------------

rs = np.linspace(1,4,1000)
N = 500
x = .5+np.zeros(N)
endcap = np.arange(round(N*.9),N)

for ri in range(len(rs)):
    
    for n in range(N-1):

        x[n+1] = rs[ri]*x[n]*(1-x[n])
        
    u = np.unique(x[endcap])
    r = rs[ri]*np.ones(len(u))
    plt.plot(r,u,'.',markersize=1,color=[(np.sin(ri/len(rs)/2)+1)/2,1-ri/len(rs),.5])


plt.title('The Logistic Map',display(Math('x_{n+1} = rx_n(1-x_n)')))
plt.grid('on')
plt.show()
```
