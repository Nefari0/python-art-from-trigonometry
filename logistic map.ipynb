```python
# exercise.independent.incomplete
r_size = 1000
n_size = 500
x = .5
n = np.linspace(1,500,n_size)
r = np.linspace(1,4,r_size)

xvector = list(range(n_size))
xvector[0] = x

for i in range(len(xvector)):
    xvector[i] = r * xvector[i] *(1-xvector[i])
    
for ploti in range(len(xvector[450:500])):
    plt.plot(r[ploti],xvector[ploti+450],'ro')

plt.show()
```
