```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display,Math
```

```python

t = np.linspace(0,15*np.pi,200)
a = t

x = a*np.cos(t)**3
y = a*np.sin(t)**3
x = np.append(x,x[0])
y = np.append(y,y[0])

plt.plot(x,y)
plt.show()
```

```python

t = np.linspace(-6*np.pi,6*np.pi,1000)

a = t

x = a*np.cos(t)**3
y = a*np.sin(t)**3

plt.plot(x,y,'m')
plt.axis('off')
plt.axis('square')
plt.show()
```
