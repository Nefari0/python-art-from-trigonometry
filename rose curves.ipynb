```python
import numpy as np
import matplotlib.pyplot as plt
```

# rose curve example:

##### x = cos(kt)cos(t)
##### y = cos(kt)sin(t)

##### t is time vector
##### k is single number

##### if k is even integer, curve will have 2k pedals
##### if k is odd integer, curve will have k pedals

```python

fig,ax = plt.subplots(3,3,figsize=(10,6))
row,col = np.indices((3,3))
t = np.linspace(0,8*np.pi,1000)

for i in range(0,9):
    
    k = i/2
    
    x = np.cos(k*t) * np.cos(t)
    y = np.cos(k*t) * np.sin(t)
    
    r = row.ravel()[i]
    c = col.ravel()[i]
    
    ax[r,c].plot(x,y,'b')
    ax[r,c].axis('off')
    ax[r,c].axis('square')
    ax[r,c].set_title('k=%s'%k)
```
